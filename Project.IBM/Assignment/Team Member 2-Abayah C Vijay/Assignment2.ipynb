Task 7 - Categorical Columns and perform Encoding


```python
data.head()
```

```python
cat_cols = [col for col in data.columns if data[col].dtype == 'object']
```

```python
le = LabelEncoder()

for col in cat_cols:
    data[col] = le.fit_transform(data[col])
```

```python
data.head()
```

Task 8 - Split the Data into Dependent and Independent Variables

```python
y = data["Balance"]
x = data.drop("Balance",axis=1)
```

```python
x
```

```python
y
```

Task 9 - Scale the Independent Variables


```python
scaled_variable = pd.DataFrame(scale(x),columns=x.columns)
scaled_variable
```

Task 10 - Split the Data into Training and Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(scaled_variable, y, test_size=0.3, random_state=0)
```
